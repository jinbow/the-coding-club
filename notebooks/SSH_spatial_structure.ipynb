# Why can't Sea Surface Height maps win a beauty contest but are still one-billon-dollar-level important? 

```python
# import ipywidgets.interact class, this class will represent a slide bar.
from ipywidgets import interact
# use interact decorator to decorate the function, so the function can receive the slide bar's value with parameter x.
```

```python
import xarray as xr
import pylab as plt
import numpy as np


@interact(x=(10,100))

def double_number(x):
    fig,ax=plt.subplots(1,2,figsize=(14,5))

    d=xr.open_dataset('~/SST_calcoast.nc')
    sst=d['SST'][0,...][:128,:128]
    del d

    cc=np.fft.fft2(sst)

    k,l=np.fft.fftfreq(128,2),np.fft.fftfreq(128,2)
    kk,ll=np.meshgrid(k,l)
    k2=kk**2+ll**2
    cc[k2>(1/x)**2]=0
    sst.data=sst.data-sst.data.mean()
    sst.plot(ax=ax[0],vmin=-0.6,vmax=0.6)
    sst.data=np.real(np.fft.ifft2(cc))
    sst.plot(ax=ax[1])
```
